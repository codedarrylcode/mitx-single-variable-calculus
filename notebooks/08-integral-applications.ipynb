## Areas and Volumes

**Areas between curves**

Given two functions or curves, $f(x), g(x)$, then the area between the two curves is equivalent to the integrand of the following:

$\text{Area} = \int_a^b (f(x) - g(x)) \cdot dx$

**Volumes by slicing**

The incremental volume of a slice of bread is as follows:

$dV = A(x) \cdot dx$

where $dx$ is a tiny slice and $A(x)$ is the cross-sectional area. $\therefore$ The volume is equivalently represented as follows:

$V = \int A(x) dx$

**Method of disks**

The method of disks gives the differential volume element for a curve rotated about the $x$-axis:

$dV = \pi y^2 dx$

<img alt="Method of Disks" src="assets/method_of_disks.png" width="600">

**Method of shells**

The method of shells for any given curve $y = f(x)$ rotated about the $y$-axis:

$dV = 2\pi x y \cdot dx$

where $2 \pi x$ is the circumference of shell and $y$ is the height.

<img alt="Method of Shells" src="assets/method_of_shells.png" width="600">

**Method of washers**

The method of washer, rotated about the $x$-axis is given by:

$dV = \pi (y_{\text{outer}}^2 - y_{\text{inner}}^2) \cdot dx$

<img alt="Method of Washers" src="assets/method_of_washers.png" width="600">

## Averages with integrals

The average value of a function, $f(x)$, between two points is the signed area under the curve, such that:

$\text{average}(f) \cdot (b-a) = \int_a^b f(x) \cdot dx$

$\therefore \text{average}(f) = \frac{1}{n} \sum_{i = 1}^{n} y_i = \frac{1}{b-a} \int_a^b f(x) \cdot dx$

This is the geometric interpretation of continuous average of a given function.

<img alt="Geometric interpretation of continuous average" src="assets/geom_interpretation_cont_average.png" width="400">

This closely relates to the Riemann sum such that, if $b-a$ is the interval of interest and $\Delta x$ are small intervals divided over $n$ intervals then:

$a = x_0 < \dots < x_n = b$

$\Delta x = \frac{b-a}{n}$

$\frac{(y_1 + \dots + y_n) \cdot \Delta x}{b-a} \to \frac{1}{b-a} \int_a^b f(x) \cdot dx \quad \text{as} \space \Delta x \to 0$

$\frac{\Delta x}{b-a} = \frac{1}{n} \to 0 \quad \text{(as} \space n \to \infty)$

$\therefore$ The average value of a function $f(x)$ on an interval $a \leq x \leq b$ is given by:

$\frac{1}{b-a} \int_a^b f(x) \cdot dx$

****

*Example 1*

Let $y(t) = -5t^2 + 2t + 11$ meters be the height of a ball. Then the velocity is $v(t) = y'(t)$. Find the average velocity over the interval $0 \leq t \leq 2$ seconds.

$\text{ave(v)} = \frac{1}{2-0} \int_0^2 y'(t) \cdot dt$

$= \frac{1}{2} (-5 - 11)$

$= -8 \space \text{meters per second}$

*Example 2*

What is the average cross-sectional area $A$ of the solid obtained by revolving the region bounded by $x = 2$, the $x$-axis and the curve $y = x^2$ about the $x$-axis?

$dV = \pi y^2 \cdot dx \space \text{(by method of disks)}$

$A = \frac{1}{2-0} \int_0^2 \pi (x^2)^2 \cdot dx = \frac{1}{2} (\frac{\pi (2)^5}{5}) = \frac{16\pi}{5}$

### MVT, FTC & Averages

The quantity $\frac{F(b) - F(a)}{b-a}$ is the average rate of change of the function $F$ over the interval $a \leq x \leq b$. 

The MVT says that there is a point $c$ such that $a < c < b$ such that $\frac{F(b) - F(a)}{b-a} = F'(c)$

By FTC1, $F(b) - F(a) = \int_a^b F'(x) dx$, thus

$\frac{F(b) - F(a)}{b-a} = \frac{1}{b-a} \int_a^b F'(x) \cdot dx$

This is also the definition of the average value of $F'$ where it is the average rate of change of $F$. In particular, $F'(c)$ is the point where $F'$ achieves its average value.
